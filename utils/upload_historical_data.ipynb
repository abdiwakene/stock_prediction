## Get data from Alpaca
200/min, delay 1s each

```python
import requests

def get_stock_data(
    symbols, 
    page_token=None,
    start='2024-03-04T00:00:00Z', 
    end='2024-03-06T00:00:00Z', 
    timeframe='1Day',
    accept='application/json',
    key_id='PKJFVW6692N7B0J3CEBY',
    secret='uzsDIdhMYFJH5R8Msk0ka4nGFyMLDlgpan7NSMtv',
):
    url = 'https://data.alpaca.markets/v2/stocks/bars'
    headers = {
        'Apca-Api-Key-Id': key_id,
        'Apca-Api-Secret-Key': secret
    }
    params = {
        'symbols': ','.join(symbols),
        'start': start,
        'end': end,
        'timeframe': timeframe
    }
    if page_token:
        params.update({'page_token': page_token})

    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        data = response.json()
        
        return data

    else:
        print("Failed to fetch data", response.status_code)
```

## Format data
- Rename cols
- Requires data col with valid str

```python
from datetime import datetime
import bson

def format_data(data_json):
    complete_df = pd.DataFrame()

    for ticker, bars in data_json['bars'].items():
        ticker_df = pd.DataFrame(bars)
        ticker_df['symbol'] = ticker
        complete_df = pd.concat([complete_df, ticker_df], ignore_index=True)

    complete_df['t'] = complete_df['t'].apply(lambda t: datetime.fromisoformat(t.replace('Z', '+00:00')))

    complete_df = complete_df.rename(columns={
        'c': 'close',
        'h': 'high',
        'l': 'low',
        'n': 'count',
        'o': 'open',
        't': 'date',
        'v': 'volume',
        'vw': 'volume_weighted'
    })
        
    return complete_df
```

## Upload docs to collection

```python
def upload_docs(data_df, client, db_name='stonks', collection_name='historical_stocks'):
    db = client[db_name]
    collection = db[collection_name]

    docs = data_df.to_dict(orient='records')

    result = collection.insert_many(docs)

    return result
```

## Iterate through SPY stocks and exec pipeline

```python
# init mongodb client 
from pymongo import MongoClient

conn_str = 'mongodb+srv://api:apipassword123@stonks.rhtscww.mongodb.net/test?retryWrites=true&w=majority'
client = MongoClient(conn_str)
```

```python
import time

for i, ticker in enumerate(tickers):
    next_page_flag = True
    page_token=None

    while next_page_flag:
        # get json
        data_json = get_stock_data(symbols=[ticker], page_token=page_token)

        # convert to df
        if len(data_json['bars']) > 0:
            data_df = format_data(data_json)

            # upload to mongodb
            result = upload_docs(data_df, client)
            print(f"Uploaded {i+1}/{len(tickers)}: {ticker}\nInserted document IDs: {result.inserted_ids}")

        next_page_flag = 'next_page_token' in data_json and data_json['next_page_token']
        page_token = data_json['next_page_token']

        # Sleep for a second to avoid hitting API rate limits
        time.sleep(1)
```
