```python
import requests

# The URL of the Flask API endpoint
url = 'http://localhost:5000/predict'

# Example data to send to the API
# Replace these values with your actual model name and features
data = {
    'symbol': 'AAPL'
}

# Make the POST request to the Flask API
response = requests.post(url, json=data)

# Check if the request was successful
if response.status_code == 200:
    # Get the prediction result from the response
    prediction = response.json()
    print("Prediction result:", prediction)
else:
    print("Error calling the API:", response.status_code, response.text)
```

```python
from datetime import datetime, timedelta
import pandas as pd

forecast = 5
last_date = datetime.now() - timedelta(days=1)
next_5 = pd.bdate_range(start=last_date, periods=forecast+1, freq='B')[1:]
```

```python
next_5
```

```python
# map forecast
retDict = {
    'forecast': []
}
for i, date in enumerate(next_5):
    retDict['forecast'].append({
        'date': date.date().strftime('%Y-%m-%dT%H:%M:%SZ')
    })
```
